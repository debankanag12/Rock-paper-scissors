```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "api_data_aadhar_enrolment_0_500000.csv"
df = pd.read_csv(file_path)
print("Dataset Loaded Successfully")
print(df.head())
print(df.info())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")


df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Remove rows with missing critical values
df = df.dropna(subset=['date'])

# Remove duplicate records
df = df.drop_duplicates()

print("Data Cleaning Completed")
print("Remaining records:", len(df))
```

```python
daily_enrolments = (
    df.groupby(['date', 'state', 'district'])
    .size()
    .reset_index(name='daily_enrolment_count')
)

print("Daily enrolment aggregation done")
```

```python
# Sort data for time-based analysis
daily_enrolments = daily_enrolments.sort_values('date')

# Calculate 7-day moving average per center (using state and district as center identifier)
daily_enrolments['moving_avg_7'] = (
    daily_enrolments
    .groupby(['state', 'district'])['daily_enrolment_count']
    .transform(lambda x: x.rolling(7, min_periods=1).mean())
)
```

```python
daily_enrolments['pct_change'] = (
    daily_enrolments
    .groupby(['state', 'district'])['daily_enrolment_count']
    .pct_change() * 100
)

print("Feature engineering completed")
```

```python
DROP_THRESHOLD = -40  # 40% drop

daily_enrolments['drop_flag'] = daily_enrolments['pct_change'] < DROP_THRESHOLD

bottleneck_centers = daily_enrolments[daily_enrolments['drop_flag'] == True]

print("Potential bottlenecks detected:")
print(bottleneck_centers[['date', 'state', 'district', 'pct_change']].head())
```

```python
center_performance = (
    daily_enrolments
    .groupby(['state', 'district'])['daily_enrolment_count']
    .mean()
    .reset_index()
    .rename(columns={'daily_enrolment_count': 'avg_daily_enrolment'})
)

# Categorize performance
center_performance['performance_label'] = pd.cut(
    center_performance['avg_daily_enrolment'],
    bins=[0, 10, 30, np.inf],
    labels=['Low', 'Medium', 'High']
)

print("Performance scoring completed")
print(center_performance.head())
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=daily_enrolments.sample(1000),
    x='date',
    y='daily_enrolment_count'
)
plt.title("Daily Aadhaar Enrolment Trend (Sample)")
plt.xlabel("Date")
plt.ylabel("Enrolments")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(
    data=center_performance,
    x='performance_label'
)
plt.title("Enrolment Center Performance Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plot_data = center_performance.sort_values('avg_daily_enrolment').head(10)
plot_data['center_identifier'] = plot_data['state'] + ' - ' + plot_data['district']
plot_data['center_identifier'] = (
    plot_data['state'] +
    plot_data['district'].astype(str) +
    ')'
)
sns.barplot(
    data=plot_data,
    x='avg_daily_enrolment',
    y='center_identifier'
)
plt.title("Top 10 Underperforming Enrolment Centers")
plt.xlabel("Average Daily Enrolment")
plt.ylabel("Enrolment Center")
plt.show()
```

```python
bottleneck_centers.to_csv(
    "identified_enrolment_bottlenecks.csv",
    index=False
)

center_performance.to_csv(
    "enrolment_center_performance_scores.csv",
    index=False
)

print("Final outputs generated:")
print("- identified_enrolment_bottlenecks.csv")
print("- enrolment_center_performance_scores.csv")
```
